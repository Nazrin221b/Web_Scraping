```python
# libraries

import requests
import openpyxl
from bs4 import BeautifulSoup
```

# Part 1 

## Collecting movie names

https://az.wikipedia.org/wiki/Az%C9%99rbaycan_tammetrajl%C4%B1_b%C9%99dii_filml%C9%99rinin_siyah%C4%B1s%C4%B1

In that website you can find names of most famous Azerbaijani movies. 
Firstly, I scraped and saved only names in order to combine them with *wikipedia link* and then collect detailed information.

```python
key_words_ulr = 'https://az.wikipedia.org/wiki/Az%C9%99rbaycan_tammetrajl%C4%B1_b%C9%99dii_filml%C9%99rinin_siyah%C4%B1s%C4%B1'

key_source = requests.get(key_words_ulr)
key_soup = BeautifulSoup(key_source.text, 'html.parser')
```

```python
film_names = key_soup.find('div', class_="mw-parser-output").find_all('li')
```

```python
key_words = []

for name in film_names:
    key = name.find('a').text
    key_words.append(key)
```

```python
key_words[:10]
```

# Part 2

## Scraping detailed movie information from wikipedia

```python
wiki_url = 'https://az.wikipedia.org/wiki/'
```

```python
# first open excel, and activate sheet.
# second step is appending column titles.

excel = openpyxl.Workbook()
sheet = excel.active
sheet.append(['Movie_name', 'Genre','Director', 'Writer', 'Painter', 'Firm','Type', 'Color', 'Runtime', 'Year' ])
```

```python
for k in key_words:

    movie_name, genre, director, writer, painter, firm, typ, color, runtime, year = '', '', '','', '', '','', '', '', '' # reset
    source = requests.get(wiki_url+k)
    soup = BeautifulSoup(source.text, 'html.parser')

    infos = soup.find('tbody')

    if infos is None:
        continue

    else:
        info = infos.find_all('tr')

        for i in info:
            th = i.find('th')
            if th in i:
                if i.find('th', class_="infobox-above title film" ):
                    movie_name = th.text


                if th.text=='Janr':
                    genre = i.find('td').get_text(strip=True, separator=', ')


                if th.text == 'Rejissor' or th.text=='Rejissorlar':
                    director = i.find('td').get_text(strip=True, separator=', ')

                if th.text == 'Ssenari müəllifi' or th.text=='Ssenari müəllifləri':
                    writer = i.find('td').get_text(strip=True, separator=', ')


                if th.text == 'Rəssam' or th.text == 'Rəssamlar':
                    painter = i.find('td').get_text(strip=True, separator=', ')

                if th.text == 'İstehsalçı':
                    firm = i.find('td').get_text(strip=True, separator=', ')

                if th.text == 'Rəng':
                    color = i.find('td').get_text(strip=True, separator=', ')

                if th.text == 'Filmin növü':
                    typ = i.find('td').get_text(strip=True, separator=', ')

                if th.text == 'Müddət':
                    runtime = i.find('td').get_text(strip=True, separator=', ').split(' ')[0]

                if th.text == 'İl':
                    year = i.find('td').get_text(strip=True, separator=', ')


        print( movie_name, genre, director, writer, painter, firm, typ, color, runtime, year)
        sheet.append([movie_name, genre, director, writer, painter, firm, typ, color, runtime, year])
```
